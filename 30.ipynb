```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df_left = DataFrame({'key': ['X', 'Y', 'Z', 'X', 'Y'], 
                     'data': range(5)})

df_left
```

```python
df_right = DataFrame({'grp_data':[10,20]}, index = ['X', 'Y'])
df_right
```

```python
pd.merge(df_left, df_right, left_on='key', right_index = True)
```
