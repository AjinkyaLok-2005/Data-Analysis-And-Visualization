```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dataset = randn(100)
```

```python
sns.displot(dataset, bins = 25, kde = True, alpha = 0.5, line_kws = {'alpha':1})
```

```python
sns.displot(dataset, rug = True, kind = 'kde')
```

```python
sns.histplot(dataset, bins = 25, color = 'blue', stat = 'density', label = 'Histogram', alpha = 0.5)
sns.kdeplot(dataset, color = 'indianred', label = 'KDE PLOT')

plt.legend()
plt.show()
```

```python
from pandas import Series

ser1 = Series(dataset, name = 'My_data')

ser1
```

```python
sns.histplot(ser1, bins = 25, alpha = 0.5)
sns.kdeplot(ser1)
```
