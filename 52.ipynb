```python
import numpy as np
import numpy.random as randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, order=4,
           scatter_kws={'marker':'o', 'color':'indianred'},
           line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,
           fit_reg = False)
```

```python
tips.head()
```

```python
tips['tips_pect'] = (tips['tip']/tips['total_bill'])*100
```

```python
tips.head()
```

```python
sns.lmplot(x='size', y='tips_pect', data=tips, col='sex', hue='smoker', aspect=1.5, markers='o')
```

```python
sns.lmplot(x='size', y='tips_pect', data=tips, x_jitter=0.1,aspect=1.5)
```

```python
sns.lmplot(x='size', y='tips_pect', data=tips, aspect = 1.5, x_estimator=np.mean)
```

```python
sns.lmplot(x='total_bill', y='tips_pect', data=tips, hue='sex', markers=['x','o'])
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tips_pect', data=tips, hue='day', aspect=1.5)
```

```python
sns.lmplot(x='total_bill', y='tips_pect', data=tips, lowess=True, line_kws={'color':'black'})
```

```python
sns.regplot(x='total_bill', y='tips_pect', data=tips)
```

```python
fig,(axis1,axis2) = plt.subplots(1, 2, sharey=True, figsize=(12, 5))

sns.regplot(x='total_bill', y='tips_pect', data=tips, ax=axis1)
sns.violinplot(x='size', y='tips_pect', data=tips, pallete='Reds', ax=axis2, aspect=1.5)
```
